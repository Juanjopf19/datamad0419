In HTML, there are a lot of messy codes such as HTML tags, Javascripts, [unicodes](https://www.w3schools.com/charsets/ref_utf_misc_symbols.asp) that will mess up your bag of words. We need to clean up those junk before generating BoW.

```python
import numpy as np
import pandas as pd
from string import punctuation
import re
```

```python
# Define your string handling functions below
# Minimal 3 functions

def strip_html_tags(text):
    notag = re.sub(".html", "", text)
    notag = re.sub("www.", "", notag)
    notag = re.sub(".com", "", notag)
    return notag

def remove_punctuation(text):
    nopunct = re.sub("\W", "", text)
    return nopunct

def to_lower_case(text):
    noupper = text.lower()
    return noupper
```

Next, paste your previously written `get_bow_from_docs` function below. Call your functions above at the appropriate place.

```python
def get_bow_from_docs(docs, stop_words=[]):
    
    corpus = []
    bag_of_words = []
    term_freq = []
    

    for doc in docs:
        file = open(doc)
        text = file.read()
        remove_punctuation(text)
        nopunct = remove_punctuation(text)
        to_lower_case(nopunct)
        noupper = to_lower_case(nopunct)
        strip_html_tags(noupper)
        notag = strip_html_tags(noupper)
        string = ''.join(notag)
        corpus.append(string)
        
    for phrase in corpus:
            words = phrase.split()
            for word in words:
                if word not in stop_words:
                    if word not in bag_of_words:
                        bag_of_words.append(word)

    for i in range(len(corpus)):
        term_freq.append([])
        words = corpus[i].split()
        for e in range(len(bag_of_words)):
            term_freq[i].append(0)
            if bag_of_words[e] in words:
                term_freq[i][e] += 1
                
    
    # Now return your output as an object
    return {
        "bag_of_words": bag_of_words,
        "term_freq": term_freq
    }
```
