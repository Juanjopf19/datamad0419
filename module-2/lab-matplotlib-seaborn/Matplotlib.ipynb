```python
import pandas as pd

df = pd.read_csv("./vehicles/vehicles.csv")

df.head()
```

```python
df.corr()
```

```python
df.groupby("Year").agg('mean').plot(y=["City MPG", "Highway MPG"])
```

```python
df.plot(x="CO2 Emission Grams/Mile", y=["City MPG", "Highway MPG"])
```

```python
%matplotlib inline
import matplotlib.pyplot as plt


plt.figure()
calc = df[["Year","Combined MPG"]].groupby("Year").agg('mean')
plt.bar(calc.index, calc["Combined MPG"])
plt.bar(calc.index, calc["Combined MPG"]*(-1))
ticks = range(1985,2015,2)
print(list(ticks))
labels = []
for t in ticks:
    if(t == 2001):
        labels.append("La ostia")
    else:
        labels.append(t)
plt.xticks(ticks, labels,rotation=90)


plt.figure()
plt.bar(calc.index, calc["Combined MPG"])
plt.legend(["Hola"])
```

```python
import numpy as np

df2 = pd.DataFrame(np.random.randn(2,50).T)

plt.figure(figsize=(12,10))

plt.subplot(2,2,1)
plt.plot(df2[0],"x")
plt.plot(df2[0],'g')
plt.plot(df2[0]+0.5, marker='o',color="#80F72E", linestyle="None")

plt.legend(["hola"])
plt.title("El primer grafico")


plt.subplot(2,2,2)
plt.title("El segundo grafico")
df2[1].plot.line()
df2[0].plot.line()
plt.legend(["hola2","hola3"])
plt.annotate("Hola",xy=(10,0),arrowprops=dict(arrowstyle='->'), xytext=(5, 3))


plt.subplot(2,2,3)
df2[0].plot.line()
plt.legend(["hola3"])

plt.subplot(2,2,4)
df2[1].plot.line()
plt.legend(["hola4"])

```
